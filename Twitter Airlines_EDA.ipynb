```python
df_negative=df[df.airline_sentiment=='negative']
df_negative.negativereason.value_counts().plot(kind='bar', color='g',figsize=(15,8), title='Distribution of negative reasons')
```

```python
df.airline.unique()
```

```python
df_Virgin_america_neg=df_negative[df.airline=='Virgin America']
df_United_neg=df_negative[df.airline=='United']
df_Southwest_neg=df_negative[df.airline=='Southwest']
df_Delta_neg=df_negative[df.airline=='Delta']
df_US_Airways_neg=df_negative[df.airline=='US Airways']
df_American_neg=df_negative[df.airline=='American']
```

```python
A=[]
A=df_negative.negativereason.value_counts().index

plt1=df_Virgin_america_neg.negativereason.value_counts()[A].plot(kind='bar',color='b',figsize=(15,15),label='Virgin America')

plt2=df_United_neg.negativereason.value_counts()[A].plot(kind='bar',color='g',
                                                         bottom=df_Virgin_america_neg.negativereason.value_counts()[A],
                                                         label="United")
plt3=df_Southwest_neg.negativereason.value_counts()[A].plot(kind='bar',color='r',
                        bottom=df_Virgin_america_neg.negativereason.value_counts()[A]+
                                                            df_United_neg.negativereason.value_counts()[A],figsize=(12,15),label='Southwest')
plt4=df_Delta_neg.negativereason.value_counts()[A].plot(kind='bar',color='c',
                                                     bottom=df_Southwest_neg.negativereason.value_counts()[A]+
                                                     df_Virgin_america_neg.negativereason.value_counts()[A]+
                                                     df_United_neg.negativereason.value_counts()[A],figsize=(12,15),label='Delta')
plt5=df_US_Airways_neg.negativereason.value_counts()[A].plot(kind='bar',color='m',
                                                          bottom=df_Delta_neg.negativereason.value_counts()[A]+
                                                          df_Southwest_neg.negativereason.value_counts()[A]+
                                                          df_Virgin_america_neg.negativereason.value_counts()[A]+
                                                          df_United_neg.negativereason.value_counts()[A],figsize=(12,15),label='US Airways')
plt6=df_American_neg.negativereason.value_counts()[A].plot(kind='bar',color='y', bottom=df_US_Airways_neg.negativereason.value_counts()[A]+
                                                        df_Delta_neg.negativereason.value_counts()[A]+
                                                        df_Southwest_neg.negativereason.value_counts()[A]+
                                                          df_Virgin_america_neg.negativereason.value_counts()[A]+
                                                          df_United_neg.negativereason.value_counts()[A],figsize=(12,15), label='American')


#blue=Virgin America
#green=United
#red=Southwest
#cyan=Delta
#magenta=US airways
#yellow=American 
```
